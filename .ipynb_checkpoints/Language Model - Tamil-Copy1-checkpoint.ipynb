# Language Model - Tamil

This project is created using Wikipedia dataset using AWD LSTM Architecture.

The Pretrained model is readily available for English, So this project will be a pretrained model for Tamil

```python
%reload_ext autoreload
%autoreload 2
%matplotlib inline

from fastai import *
from fastai.text import *
```

```python
bs=4
# torch.cuda.set_device(2)
data_path = Config.data_path()

lang = 'ta'
name = f'{lang}wiki'
path = data_path/name
# path.mkdir(exist_ok=True, parents=True)
```

```python
mdl_path = path/'models'
# mdl_path.mkdir(exist_ok=True)
lm_fns = [mdl_path/f'{lang}_wt', mdl_path/f'{lang}_wt_vocab']
```

## Tamil wikipedia model

Utility Program downloads and stores wikipedia data

```python
from nlputils import split_wiki,get_wiki

get_wiki(path,lang)
!head -n4 {path}/{name}
```

```python
path
```

```python
# dest = split_wiki(path,lang)
dest = path/'docs'
```

274392 Unique Vocabulary is identified. Here am using FastAI datablock API which splits the data into Train and Valid Dataset.

If a word doesn't repeat for 5 times in dataset then it is denoted as xxunk.

Batch size also maintained such that it can pass those data into model batch wise.

```python
data = (TextList.from_folder(dest)
        .split_by_rand_pct(0.1, seed=42)
        .label_for_lm()
        .databunch(bs=bs, num_workers=1))

data.save(f'{lang}_databunch')
len(data.vocab.itos),len(data.train_ds)
```

```python
data = load_data(dest, f'{lang}_databunch', bs=bs)
```

```python
data.vocab.itos
```

```python
data.show_batch()
```

```python
learn = language_model_learner(data, AWD_LSTM, drop_mult=0.1, wd=0.1, pretrained=False).to_fp16()
```

```python
lr = 3e-3
lr *= bs/4  # Scale learning rate by batch size
```

```python
learn.unfreeze()
learn.fit_one_cycle(10, lr, moms=(0.8,0.7))
```

```python
learn.predict('காதல் ஆசை', 50, 0.6)
```

```python
learn.to_fp32().save(lm_fns[0], with_opt=False)
learn.data.vocab.save(lm_fns[1].with_suffix('.pkl'))
```
